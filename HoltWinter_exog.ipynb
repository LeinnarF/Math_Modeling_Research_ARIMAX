```python
import warnings
warnings.filterwarnings("ignore")

from utils import *

df = pd.read_csv('dataset.csv', index_col='Month', parse_dates=True,date_parser=lambda x: pd.to_datetime(x, format='%y-%b'))

production = df['Production']
withdrawals = df['Withdrawals']
globalPrice = df['GlobalPrice']
exchangeRate = df['ExchangeRate']
temperature = df['TempMax']
precipitation = df['Precipitation']
```

# Production

```python
prod_train, prod_test = train_test_split(production, 24)

pred = HoltWintersModel(prod_train, prod_test)
Evaulate_HW(prod_test, pred)
prod_forecast = HoltWintersForecast(production, 36)
HWForecast_plot(production, prod_forecast)
```

# Withdrawals

```python
with_train, with_test = train_test_split(withdrawals, 24)

pred = HoltWintersModel(with_train, with_test)
Evaulate_HW(with_test, pred)
with_forecast = HoltWintersForecast(withdrawals, 36)
HWForecast_plot(withdrawals, with_forecast)
```

# Global Price

```python
gPrice_train, gPrice_test = train_test_split(globalPrice, 24)
pred = HoltWintersModel(gPrice_train, gPrice_test)
Evaulate_HW(gPrice_test, pred)
gPrice_forecast = HoltWintersForecast(globalPrice, 36)
HWForecast_plot(globalPrice, gPrice_forecast)
```

# Exchange Rate

```python
eRate_train, eRate_test = train_test_split(exchangeRate, 24)
pred = HoltWintersModel(eRate_train, eRate_test)
Evaulate_HW(eRate_test, pred)
eRate_forecast = HoltWintersForecast(exchangeRate, 36)
HWForecast_plot(exchangeRate, eRate_forecast)
```

# Temperature

```python
temp_train, temp_test = train_test_split(temperature, 24)
pred = HoltWintersModel(temp_train, temp_test)
Evaulate_HW(temp_test, pred)
temp_forecast = HoltWintersForecast(temperature, 36)
HWForecast_plot(temperature, temp_forecast)
```

# Precipitation

```python
precip_train, precip_test = train_test_split(precipitation, 24)
pred = HoltWintersModel(precip_train, precip_test)
Evaulate_HW(precip_test, pred)
precip_forecast = HoltWintersForecast(precipitation, 36)
HWForecast_plot(precipitation, precip_forecast)
```

# Combine

```python
# Combine all the forecasts into a single DataFrame

combined_forecasts = pd.DataFrame({
    'Production': prod_forecast,
    'Withdrawals': with_forecast,
    'GlobalPrice': gPrice_forecast,
    'ExchangeRate': eRate_forecast,
    'TempMax': temp_forecast,
    'Precipitation': precip_forecast
})

combined_forecasts.to_csv('HWForecasts.csv', header=True, index=True)
```
